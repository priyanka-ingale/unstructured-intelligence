2. find cumulative density **cdf** using `cumsum()`
3. normalize **cdf** to make sure we are in the range [0, 255]
4. `interp()` interpolate the image array using **cdf** to generate normalized image which has piece-wise linear cumulative distribution
5. reshape the normalized image
